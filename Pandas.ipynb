```python
df=df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
```

```python
df
```

```python
df=df.drop_duplicates()
```

```python
df
```

```python
df.isnull().sum()
```

```python
data=df.dropna(subset=['continent'])
```

```python
data
```

```python
data.isnull().sum()
```

```python
df=data.fillna(0)
```

```python
df.isnull().sum()
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
df['Month']=df['date'].dt.month
```

```python
df
```

```python
df_groupby=df.groupby('continent')
```

```python
df_groupby=df_groupby.max()
```

```python
df_groupby=df_groupby.reset_index() 
```

```python
df_groupby
```

```python
df_groupby['total_deaths_to_total_cases']=df_groupby['total_deaths']/df_groupby['total_cases']
```

```python
df_groupby
```

```python
sns.distplot(df_groupby['gdp_per_capita'])
```

```python
sns.jointplot(data=df_groupby,x='total_cases',y='gdp_per_capita',kind='scatter')
```

```python
sns.pairplot(data=df_groupby)
```

```python
sns.catplot(data=df_groupby,x='continent',y='total_cases',kind='bar')
```
